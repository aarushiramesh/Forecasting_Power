# Reduced Power consumption data frame

```python
df_ercot = pd.read_csv('ercot_combined.csv')
```

```python
df_reduced_ercot = df_ercot.drop(columns=['COAST', 'EAST', 'WEST', 'FWEST', 'NORTH', 'NCENT', 'SOUTH', 'WEST', 'Time', 'ERCOT', 'Year', 'Month', 'Day', 'Unnamed: 0'])
df_reduced_ercot
```

```python
df_reduced_ercot[['Month','Day', 'Year']] = df_reduced_ercot['Date'].astype(str).str.split("/", 3, expand=True)
df_reduced_ercot = df_reduced_ercot.drop(['Date'], axis=1)
df_reduced_ercot['Date'] = df_reduced_ercot['Month'] + "/" + df_reduced_ercot['Day'] + "/"+ df_reduced_ercot['Year']
df_reduced_ercot
```

```python
df_reduced_ercot = df_reduced_ercot.groupby(['Date'], as_index=False).mean()
df_reduced_ercot
```

## Reducing population Data Frame

```python
df_pop = pd.read_csv('pop_features_cleaned.csv')
df_pop = df_pop.drop(columns=['Unnamed: 0'])
df_reduced_pop = df_pop[df_pop['Region'] == 'SCENT']
df_reduced_pop
```

```python
df_reduced_pop = pd.read_csv("manual_pop.csv")
df_reduced_pop
```

```python
df_weather = pd.read_csv('Austin_Weather.csv')
df_weather = df_weather.drop(columns=['Unnamed: 0'])
```

```python
df_weather['PrecipiationTotalInches'] = df_weather['PrecipitationSumInches'].replace(['T'],'0')
df_weather_precip = df_weather.drop(columns='PrecipitationSumInches')
```

```python
df_weather_precip['WindAvgMPH'] = df_weather_precip['WindAvgMPH'].replace(['-'], 0)
df_weather_precip['WindAvgMPH'] = df_weather_precip['WindAvgMPH'].astype(int)
mean_wind = df_weather_precip['WindAvgMPH'].mean()
df_weather_precip['WindAvgMPH'] = df_weather_precip['WindAvgMPH'].replace([0], mean_wind)
```

```python
df_weather_precip[['Year','Month', 'Day']] = df_weather_precip['Date'].astype(str).str.split("-", 3, expand=True)
df_weather_precip = df_weather_precip.drop(['Date'], axis=1)
df_weather_precip['Date'] = df_weather_precip['Month'] + "/" + df_weather_precip['Day'] + "/"+ df_weather_precip['Year']
df_weather_precip
```
